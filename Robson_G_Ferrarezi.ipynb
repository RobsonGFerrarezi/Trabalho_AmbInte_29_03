Agora vamos pegar as informações de postos de São Paulo, e Santa Catarina acima de R$ 2,15 para o ETANOL

```python
filter5 = (((df['ESTADO'] == 'SAO PAULO') | (df['ESTADO'] == 'SANTA CATARINA')) & ((df['PRODUTO'] == 'ETANOL HIDRATADO') & (df['PREÇO MÉDIO REVENDA'] > 2.15)))
df[filter5]
```

Como se pode ver o processo foi feito como pedido, e filtrado, trazendo 815 linhas com os paramêtros passados.



---

Agora vamos passar a tabela para um arquivo CSV salvo no Drive.

```python
df[filter5].to_csv('drive/MyDrive/Colab Notebooks/dados/minha_pesquisa')
```
